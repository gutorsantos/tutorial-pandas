#### Método loc

Método importando, a partir dele é possível realizar:
+ seleção de uma linha
+ seleção de linhas dada condição
+ seleção de linhas e colunas específicas
+ seleção de um valor específico de uma linha

```python
# selecao de uma linha
df_train.loc[1]
```

```python
# selecao de linhas dada condicao
df_train.loc[df_train['PoolArea'] > 0]
```

```python
# selecao de linhas e colunas especificas
df_train.loc[df_train['PoolArea'] > 0, ['Id', 'PoolArea', 'SalePrice']]
```

```python
# selecao de um valor especifico de uma linha
df_train.loc[0, 'SalePrice']
```

#### Adicionar uma coluna

```python
df_train['Comissao'] = df_train['SalePrice'] * 0.05
df_train.head()
```

#### Remover uma coluna

```python
df_train = df_train.drop('Comissao', axis=1)
```

#### Valores vazios

Para trabalhar valores vazios, podemos realizar:
+ remoção linhas e colunas completamente vazias
+ remoção linhas que possuem valores vazios
+ preencher valores vazios

```python
# remocao de linhas e colunas completamente vazias
removed = df_train.dropna(how='all', axis=0)
removed = df_train.dropna(how='all', axis=1)

removed.tail()
```

```python
# remocao de linhas e colunas com pelo menos um valor vazia
removed = df_train.dropna()         # (how='any', axis=0)
removed = df_train.dropna(axis=1)   # (how='any')

removed.tail()
```

```python
# preencher valores vazios
df_train['LotFrontage'] = df_train['LotFrontage'].fillna(df_train['LotFrontage'].mean())    # preenchendo com os valores médios
df_train.loc[1446, ['LotFrontage']]
```

### Método _value_counts()_

Seleciona-se uma coluna, o pandas automaticamente identifica todos os valores distintos e realiza a contagem da quantidade de ocorrências.



```python
pool_area = df_train['PoolArea'].value_counts()
pool_area
```

### Método _groupby()_

Passa-se como argumento uma coluna a qual será o parâmetro que irá agrupar as linhas, após disso é possível selecionar um método de agrupamento.

Exemplo, selecionando o YrSold como parâmetro de agrupamento, e, em seguida, seja escolhido o sum, será realizada a soma dos valores numéricos da linhas agrupadas. Portanto, para a coluna SalePrice podemos ver o montante acumulado de todas as vendas realizadas para o ano X.

```python
# year_sold = df_train.groupby('YrSold').min()
# year_sold = df_train.groupby('YrSold').max()
# year_sold = df_train.groupby('YrSold').mean()
# year_sold = df_train.groupby('YrSold').count()
year_sold = df_train.groupby(['YrSold']).sum()
year_sold
```
